#### Statistical Testing

```python
numOfAlpha = 0
numOfGamma = 0
numOfBeta = 0
numOfOmega = 0
for index, item in df.iterrows():
  if item['Placement'] == "Yes" and item['Internship_Experience'] == 'Yes':
    numOfAlpha += 1
  elif item['Internship_Experience'] == "Yes" and item['Placement'] == 'No':
    numOfGamma += 1
  elif item['Internship_Experience'] == "No" and item['Placement'] == 'Yes':
    numOfBeta += 1
  else:
    numOfOmega += 1

print(f"""
# of students who were placed and had an internship: {numOfAlpha}
# of students who were placed and had NO internship: {numOfBeta}
# of students who were NOT placed and had an internship: {numOfGamma}
# of students who were NOT placed and had NO internship: {numOfOmega}

% of interns who are placed: {numOfAlpha / (numOfAlpha + numOfGamma) * 100}%
% of interns who are NOT placed: {numOfGamma / (numOfAlpha + numOfGamma) * 100}%

% of placed students who are interns: {numOfAlpha / (numOfAlpha + numOfBeta) * 100}%
% of placed students who are NOT interns: {numOfBeta / (numOfAlpha + numOfBeta) * 100}%
""")

placementvalues = placed['Internship_Experience'].value_counts().sort_values()
plt.pie(placementvalues, labels=['Placed with internship, 39%', 'Placed without internship, 61%'])
plt.title('Proportion of placed students with and without internships.')
plt.show()
```
